```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("supermarket.csv", header=None)

plt.stem(data[2])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("supermarket.csv", header=None)

plt.bar(data[0], data[2])

plt.title("Supermarket Price")
plt.xlabel("Bill Number")
plt.ylabel("Price")

plt.show()
```

```python
plt.scatter(data[1], data[2])

plt.title("Quantity vs Price")
plt.xlabel("Quantity")
plt.ylabel("Price")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("supermarket.csv", header=None)

plt.pie(data[2], labels=data[0], autopct="%1.1f%%")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("supermarket.csv", header=None)

sns.violinplot(y=data[2])
plt.title("Violin Plot")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("supermarket.csv", header=None)

sns.kdeplot(data[2])

plt.title("KDE Plot")
plt.xlabel("Price")
plt.show()
```
