```python
# 9. Matrix Operations
# QUESTION:1 - Create a 3 × 3 matrix.
c=np.arange(2,11)
print(c)
```

```python
w=np.reshape(c,(3,3))
```

```python
#QUESTION:2 -  Find its transpose.

np.transpose(c)
```

```python
# QUESTION:3 - Find the sum of each row.

#w[ s :e ,  s:e] 
print(w)

print("Row 1 sum:",w[0:1].sum())
print("Row 2 sum:",w[1:2].sum())
print("Row 3 sum:",w[2:3].sum())
```

```python
# QUESTION:4 - Find the sum of each column

print("column 1 sum:",w[:,0].sum())
print("column 2 sum:",w[:,1].sum())
print("column 3 sum:",w[:,2].sum())
```
